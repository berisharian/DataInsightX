```sql
SELECT COUNT(sod.ProductID) as Quantity, sod.ProductID, p.[Name], SUM(sod.LineTotal) as TotalSale
FROM Sales.SalesOrderDetail sod 
INNER JOIN Production.Product p on p.ProductID=sod.ProductID
GROUP BY sod.ProductID, p.[Name]
HAVING COUNT(sod.ProductID)>1250
ORDER BY COUNT(sod.ProductID) DESC
```

```sql
Select t.TerritoryID, [Name], CountryRegionCode, [Group], CustomerCount
FROM (
    SELECT c.TerritoryID, COUNT(*) AS CustomerCount
    FROM Sales.Customer c 
    GROUP BY c.TerritoryID) as t
INNER JOIN Sales.SalesTerritory st ON st.TerritoryID=t.TerritoryID
ORDER BY CustomerCount
```

```sql
SELECT
    [Name],
    AVG(CASE WHEN DaysToManufacture = 0 THEN SalePrice END) AS AvgDays0,
    AVG(CASE WHEN DaysToManufacture = 1 THEN SalePrice END) AS AvgDays1,
    AVG(CASE WHEN DaysToManufacture = 2 THEN SalePrice END) AS AvgDays2,
    AVG(CASE WHEN DaysToManufacture = 3 THEN SalePrice END) AS AvgDays3,
    AVG(CASE WHEN DaysToManufacture = 4 THEN SalePrice END) AS AvgDays4
FROM (
    SELECT
        pc.[Name],
        sod.UnitPrice * (1 - sod.UnitPriceDiscount) AS SalePrice,
        p.DaysToManufacture
    FROM
        Production.ProductCategory pc
        INNER JOIN Production.ProductSubCategory psc ON pc.ProductCategoryID = psc.ProductCategoryID
        INNER JOIN Production.Product p ON p.ProductSubCategoryID = psc.ProductSubCategoryID
        INNER JOIN Sales.SalesOrderDetail sod ON sod.ProductID = p.ProductID
) AS dm
GROUP BY [Name];
```

```sql
CREATE TRIGGER trg_excessInventory ON Production.ProductInventory
FOR UPDATE
AS 
BEGIN 
    IF EXISTS (
        SELECT 'True'
        FROM inserted i 
        JOIN deleted d on i.ProductID=d.ProductID AND i.LocationID=d.LocationID
        WHERE i.Quantity>800
    )
    BEGIN 
        RAISERROR('Cannot increase stock where units would be over 800 units', 16, 1);
        ROLLBACK TRAN
    END
END;
```

```sql
CREATE TRIGGER trg_storeModifiedDate ON Sales.Store 
AFTER INSERT, UPDATE 
AS 
BEGIN 
    UPDATE s
    SET ModifiedDate=GETDATE()
    FROM inserted i 
    INNER JOIN Sales.Store s ON s.BusinessEntityID=i.BusinessEntityID
END;
```

Problem 9: Using transactions insert a new record to Purchasing.ShipMethod, update another existing record and then commit the changes if there is no error occured, otherwice rollback to maintain data consistency.

```sql
SELECT *
FROM Purchasing.ShipMethod
```

```sql
--SET IDENTITY_INSERT Purchasing.ShipMethod ON;
BEGIN TRANSACTION T1
    INSERT INTO Purchasing.ShipMethod ([Name], ShipBase, ShipRate, rowguid, ModifiedDate) 
    VALUES ('New Ship Method Name 2', 8.8, 1.8, NEWID(), GETDATE());

    UPDATE Purchasing.ShipMethod 
    SET ShipRate = 2.01
    WHERE ShipMethodID=1

    IF(@@ERROR > 0) 
    BEGIN 
        ROLLBACK TRANSACTION
        PRINT 'Transaction rolled back!'
    END 
    ELSE 
COMMIT TRANSACTION T1
--SET IDENTITY_INSERT Purchasing.ShipMethod OFF;
```
